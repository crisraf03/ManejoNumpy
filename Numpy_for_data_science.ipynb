# Capitulo 2.4: Pseudo-random generación de números

Dentro de la libreria numpy existe la seccion random, esta es la que vamos a usar. Para ello vamos a importarla de forma particular y etiquetarla para mayor comodidad

```python
import numpy.random as r

        'probemos usar la funcion randn la cual genera numero "aleatorios"''
s1 = r.randn()
s2 = r.randn(3)
s3 = r.randn(2,3)
s4 = r.randn(3,2,2)

print('1. diferentes dimensiones y formas de arreglos con numeros aleatorios: ' )
print('s1 = ', s1,'s2 = ',  s2,'s3 = ',s3)
print('s4 = ',s4)

        '''se se desea se puede escoguer valores random de la distribucion gauseana, la campana de gauss'''
s5 = r.normal(size = (4,2)) # le agregamos las dimensiones deseas en este caso 
print('2. Matriz de valores dist. Normal: ', s5)
r.seed(1234)#por averiguar que hace la funcion r.seed()

        '''podemos tambien generar numeros enteros aleatorios usando randint'''
s6 = r.randint(5, 10) # se le introduce los limites sobre los cuales tomara dicho numero pseduo aleatorio
print ('3. entero aleatorio generado:' , s6)
 
        '''podemos hacer un reshape a un conjunto de datos para transformarlo a la forma que queramos'''
s7 = r.randn(2,3)
print('4. transformar un conjunto de datos aleatorios a la forma que deseemos :')
print(s7 ,s7.reshape(1,6))

        '''podemos crear arreglos de numeros desde un arreglo inicialmente aleatorio'''
arreglo_aleatorio = r.RandomState(1234)
s8 = arreglo_aleatorio.rand(4) # arreglo random
s9 = arreglo_aleatorio.randint(10) #numero entero random
s10 = arreglo_aleatorio.normal(10) #desde la distribucion normal
s11 =  arreglo_aleatorio.uniform(10) #desde una distribucion uniforme (unidad)

print ('5. los diferentes tipos de arreglos aleatorios que se pueden usar una a partir de un arreglo aleatorio inicial, a cada compilacion daran nuevos valores.')
print ('s8: ', s8 )
print('s9: ', s9) 
print('s10: ', s10)
print ('s11: ', s11)
```

# seccion 3: Indexación en numpy

## Capitulo 3.1: Indexing and Slicing 



```python
        'Creamos un nuevo arreglo y utilizamos la indexación para seleccionar un elemento de está'

sc31 = np.array(np.arange(10))
print('1. nuevo arreglo sobre el cual se trabajara:' , sc31)
print('2. sc3_1 elemento en la posición 3: ', sc31[3])

        'se puede tambien seleccionar una seccion de este arreglo limitando las posiciones seleccionadas'
print('3. seleccionando un rango de limites: ', sc31[2:6])
        
        'puedes reescribir los elementos en un arreglo igualandolo a un valor concreto'
sc31[2:6] = 5
print('4. arreglo de datos con introduccion del 5 en las posiciones [2:6] : ', sc31)

        'Además a partir de la selección de un grupo de datos podemos crear un nuevo arreglo'
sc32 = sc31[2:6]
print('5. nuevo arreglo a partir de la selección de sc31: ', sc32)

        'se puede cambiar el valor de un elemento en especifico por otro, asi como de todos los valores de esta'
sc31[5] = 50
print('6. cambiando un valor de arreglo "sc31[5] = 50" :', sc31)
sc32[:] = 20
print('7. Cambiando todos los elementos de un arreglo "sc32[:] = 20" : ', sc32)

        'creando un nuevo arreglo de 2 dimensiones; se puede seleccionar elementos pero habrá que seleccionar sus coordenadas [i][j]'
sc33 = np.arange(9).reshape(3,3)
print('8. arreglo 2dim inicial sc33:' )
print(sc33)
print ('9. Se seleciona las posiciones de un elemento en 1 dim y en 2 dim: ')
print('sc33[2] =  ',sc33[2],'/ sc33[2][1] = ' ,sc33[2][1])

        'Tambien se puede seleccionar una posición de forma más comoda usando comas A[a,b]'
print('10.  usando la sintaxis A[a,b] : sc33[2,1] = ', sc33[2,1])

        'De forma analoga para 3 o mas dimensiones'
    
sc34 = np.arange(12).reshape(2,2,3)
print('11. matriz de 3 dimensiones:')
print(sc34)
print('12. Seleccionando el elemento A[a,b,c]: sc34[0,1,2] =  ', sc34[0,1,2])

        'con el método np.copy() podemos copiar un arreglo en otro'
sc35 = sc34[0,0].copy()
print('13. Usando np.copy() para copiar un arreglo, copiamos una parte de la matriz sc34: ', sc35)


        'Se aplican las mismas operaciones vistas anteriormente de reemplazo de elementos en los arreglos'
sc34[0] = 100
print('14. Reemplazando valores sc34[0] = 100 : ')
print(sc34)


        'se puede limitar un maximo o minimo para la seleccion de elementos incluso si estos son arreglos dentro de otros '
print ( '15. seleccion sc34[1:,1]  :')
print(sc34[1:,1]) # la seleccion se hace de la siguiente forma [ inf:sup ,inf:sup , inf:sup , ...] dependiendo de las dimensiones del arreglo.
print('16. De una forma mas interesante se puede seleccionar solo un recaudro del arreglo (area, volumen, n-volumen): ')
print('de la siguiente forma sc34[1:,:1]')
print (sc34[1:,:1])

        'Veamos un poco mas el PODER  de esta herramienta'
sc36 = np.arange(125).reshape(5,5,5) # cubo de datos de arista 5
    #se selecciona un pedazo de este volumen
print ('17. volumen de datos de arista 5: ' )
print (sc36) # mirando la correspondencia de los datos se puede ver un  cubo laminar L = 5
print ('seleccionamos unicamente un cubo de arista 3')
print (sc36[:3,:3,:3]) # se recortan de las laminas otras mas pequeñas de arista L = 3
print('puede ser incluso un hilo de este cubo: ')
print(sc36[0,:,0]) # realmente impresionante ahora que por primera vez lo voy entendiendo

        'probemos ahora cambiar los valores de una parte de este cubo'
    
sc36[:,1:4,1:4] = 25
print('18. Cambiando los valores de una parte de este cubo: ')
print(sc36)

```

## Capitulo 3.2: boolean indexing

Podemos ingresar indexacion a partir de elementos booleanos a pesar que la indexacion en numpy solo acepte directamente enteros.

```python
        'Aprender el uso de indexación con string'

colores = np.array(['rojo', 'azul','verde' , 'rojo', 'azul','verde' ,'rojo', 'azul','verde','rojo']) # creamos un arreglo con elementos string
datos = np.arange(40).reshape(10,4) #creamos una matriz con elementos 
print('1. Mostramos la matriz de datos creadas: ')
print(datos)

        'eleccionamos el valor booleano de cada elemento del arreglo colores'
# es fundamental que el arreglo de seleccion y el arreglo de elementos (a seleccionar) tengan las mismas dimensiones
print('2. selecion de valor booleano de colores == rojo :', colores == 'rojo') 
        
        'seleccionar a partir de valores booleanos el elemento de un arreglo'
print('3. datos [colores == rojo] : ')
print(datos[colores=='rojo']) # para este caso se selecciona un elemento de dim (4,)

        'escojamos ahora los elementos de esta nueva matriz'
print('4. escoguiendo un sub matriz de está: datos[colores == rojo, 1:]')
print(datos[colores == 'rojo',1:])
print('5. seleccionando elementos')
print(datos[colores == 'rojo', 1] )

        'si se desea negar la condición se usa !=  o ~ '
print('6. negando la condicion booleana anterior datos[colores != rojo ]: ')
print(datos[colores != 'rojo' ])

        'si se desea tomar más de una condición de puede usar el simbolo |'
print('7. usar una condicion u otra')
condiciones = (colores == 'rojo') | (colores == 'azul')
print(datos[condiciones])

        'se  puede usar una desigualdad como condicion de seleccion (muy util)'
print('8. arreglo de numeros para usar como condicion de seleccion :')
numeros = np.arange(10)

print('9. datos con la seleccion de lista numeros menores estrictos a 8:')
print(datos[numeros <8])

# ademas se puede seleccionar por ambos lados 
print('10. y seleccion de estos desde el elemento 0 al 2 de cada fila:')
print(datos[numeros<8, :2 ])

        'se pueden usar los propios valores del arreglo como criterio de selecion '
    #EL PROPIO ARREGLO! prácticamente es aprender a hablar y pedirle las cosas comodamente
print('11. seleccionando solo aquellos datos menores a 25: ')
print(datos[datos<25])

        'finalmente podemos no solo seleccionar si no reemplazar todos estos valores de un arreglo n dim por otros dados manualmente'
print('12. reemplazando datos[datos < 30] = 7: ')
datos[datos<30] = 7
print(datos)

```
